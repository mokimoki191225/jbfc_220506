```python

import numpy as np
import matplotlib.pyplot as plt

def plot_slope(X, Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys/Xs)
    
X = np.linspace(-3, 3, 100)
Y = np.exp(-X**2)

plt.plot(X, Y)
plot_slope(X, Y)
plt.show()
```

```python
# 값을 두 집합 이상으로 분할
import numpy as np
import matplotlib.pyplot as plt

A = np.array([5., 30., 45., 22.])
B = np.array([5., 25., 50., 20.])
C = np.array([1.,  2.,  1.,  1.])
X = range(4)

plt.bar(X, A, color='b')
plt.bar(X, B, color='g', bottom=A)
plt.bar(X, C, color='r', bottom=A+B)
plt.show()
```

```python
# 값을 두 집합 이상으로 분할
import numpy as np
import matplotlib.pyplot as plt

data = np.array([[5., 30., 45., 22.],
                 [5., 25., 50., 20.],
                 [1.,  2.,  1.,  1.]        
    ])

color_list = ['b', 'g', 'r']

X = np.arange(data.shape[1])

for i in range(data.shape[0]):
    plt.bar(X, data[i],
           bottom = np.sum(data[:i], axis=0),
           color = color_list[i % len(color_list)])

plt.show()
```

```python
# 두 막대 차트를 동시에 양방향으로 표시
import numpy as np
import matplotlib.pyplot as plt

women_pop = np.array([5., 30., 45., 22.])
men_pop   = np.array([5., 25., 50., 20.])
X = np.arange(4)

plt.barh(X, women_pop, color='r')
plt.barh(X, -men_pop, color='b')
plt.xticks([-10,-5,0,5,10],('10','5','0','5','10'))    # Get or set the *x*-limits of the current tick locations
plt.show()
```

```python
# 공간 위치를 처리할 때 삼각화 생긴다. 점과 이웃 관계 사이의 거리를 보여줌
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri

data = np.random.rand(100, 2)

triangles = tri.Triangulation(data[:,0], data[:,1])

plt.triplot(triangles)
plt.show()
```

```python
# 여러 수량을 비교
import numpy as np
import matplotlib.pyplot as plt

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

X = np.arange(4)
plt.bar(X+0.00, data[0], color='b', width=0.25)
plt.bar(X+0.25, data[1], color='g', width=0.25)
plt.bar(X+0.50, data[2], color='r', width=0.25)

plt.show()
```

```python
# 반복자 enumerate 는 현재 행과 해당 첨자를 둘 다 반환
import numpy as np
import matplotlib.pyplot as plt


data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

color_list = ['b','g','r']
gap = .8/len(data)
for i, row in enumerate(data):
    X = np.arange(len(row))
    plt.bar(X + i*gap, row, width = gap, color = color_list[i%len(color_list)])

plt.show()
```
