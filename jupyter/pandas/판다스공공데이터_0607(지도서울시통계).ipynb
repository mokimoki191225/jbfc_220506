```python
population_df = pd.read_table(population_txt, header=2)
population_df.head()
```

```python
population = population_df[['자치구', '세대', '계', '세대당인구', '65세이상고령자']]
population.head()
```

```python
tmp = population['세대'][0]
tmp = tmp.replace(',', '')
tmp
```

```python
population.iloc[0]['세대']
```

```python
population.index
```

```python
for idx in population.index:
    # print(population['계'][idx].replace(',', ''))
    tmp = population['세대'][idx]
    tmp = tmp.replace(',', '')
    # print(tmp)
    population['세대'][idx] = tmp
```

```python
for idx in population.index:
    # print(population['계'][idx].replace(',', ''))
    tmp = population['65세이상고령자'][idx]
    tmp = tmp.replace(',', '')
    population['65세이상고령자'][idx] = tmp
```

```python
for idx in population.index:
    # print(population['계'][idx].replace(',', ''))
    tmp = population['계'][idx]
    print(tmp, end='  \t-->  ')
    tmp = tmp.replace(',', '')
    # tmp = tmp.strip()
    print(tmp, type(tmp))
    population['계'][idx] = tmp
```

```python
import json
import folium
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
geo_path = 'data/skorea_municipalities_geo_simple.json'
```

```python
geo_str = json.load(open(geo_path, encoding='utf-8'))
# geo_str
```

```python
guDat = pd.DataFrame({'gu':population['구'], 'counts':population['인구수']})
guDat.head()
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')
map
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')

map.choropleth(geo_data=geo_str,
              data=guDat,
              columns=['gu', 'counts'],
              fill_color='PuRd', #PuRd, YlGnBu
              key_on='feature.id')
```
