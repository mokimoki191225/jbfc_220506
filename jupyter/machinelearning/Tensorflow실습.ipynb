```python
import tensorflow.compat.v1 as tf

tf.disable_v2_behavior()
```

```python
a = tf.Variable(5)
b = tf.Variable(3)
c = tf.multiply(a,b)

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

print(sess.run(c))
```

```python
sess = tf.Session()

x = tf.Variable(tf.random_normal([1,356], mean = 1, stddev=0.35))

# 356개의 열을 1개만드는 random_normal
init = tf.global_variables_initializer()

sess.run(init)

print(sess.run(x))
```
